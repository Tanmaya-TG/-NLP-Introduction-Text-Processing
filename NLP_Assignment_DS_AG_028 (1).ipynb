## Question 1: What is Computational Linguistics and how does it relate to NLP?

**Answer:**
Computational Linguistics is an interdisciplinary field that focuses on the computational modeling and analysis of human language. It combines concepts from linguistics, computer science, and artificial intelligence to enable computers to process and understand natural language.

Natural Language Processing (NLP) is a practical application of computational linguistics. While computational linguistics focuses more on theoretical models of language, NLP focuses on implementing these models in real-world systems. NLP uses computational linguistic principles to build applications such as chatbots, machine translation systems, voice assistants, sentiment analysis tools, and text summarization systems.

## Question 2: Briefly describe the historical evolution of Natural Language Processing.

**Answer:**
The development of NLP has gone through several phases:

1. **1950s – Early Beginnings:**
   Research began with machine translation projects. The famous Turing Test by Alan Turing also influenced language understanding in machines.

2. **1960s–1970s – Rule-Based Systems:**
   NLP systems relied on hand-written grammar rules and linguistic knowledge. Systems like ELIZA were developed during this time.

3. **1980s–1990s – Statistical NLP:**
   With increased computing power and digital text, statistical models such as Hidden Markov Models and probabilistic approaches became popular.

4. **2000s – Machine Learning Era:**
   Algorithms like Naive Bayes, Support Vector Machines, and decision trees were used for tasks like classification and sentiment analysis.

5. **2010s–Present – Deep Learning Era:**
   Neural networks, word embeddings (Word2Vec, GloVe), and transformer models like BERT and GPT significantly improved NLP performance.

## Question 4: What is text normalization and why is it essential in text processing tasks?

**Answer:**
Text normalization is the process of converting text into a standard and consistent format so that it can be easily processed by machines.

Common normalization techniques include:
- Converting text to lowercase
- Removing punctuation
- Removing stop words
- Stemming and lemmatization
- Expanding contractions

Text normalization is important because raw text often contains inconsistencies such as different letter cases, punctuation, or variations of the same word. Normalizing the text improves the quality of data and helps machine learning models perform better.

## Question 5: Compare and contrast stemming and lemmatization with suitable examples.

**Answer:**

| Feature | Stemming | Lemmatization |
|--------|---------|---------------|
| Definition | Removes word endings to get root form | Converts words to their dictionary base form |
| Accuracy | Less accurate | More accurate |
| Method | Rule-based truncation | Uses vocabulary and morphological analysis |
| Example | 'running' → 'run' or 'runn' | 'running' → 'run' |

Example:
- Words: *running, runner, runs*
- Stemming result: *run*
- Lemmatization result: *run*

## Question 6: Extract email addresses using Regular Expressions

```python
import re

text = """Hello team, please contact us at support@xyz.com for technical issues, or reach out to
our HR at hr@xyz.com. You can also connect with John at john.doe@xyz.org and jenny
via jenny_clarke126@mail.co.us. For partnership inquiries, email partners@xyz.biz."""

pattern = r'[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[a-zA-Z]+'

emails = re.findall(pattern, text)

print("Extracted Emails:")
for email in emails:
    print(email)
```

## Question 7: Tokenization and Frequency Distribution using NLTK

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

nltk.download('punkt_tab') # Changed from 'punkt' as per error suggestion

text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence."""

tokens = word_tokenize(text)
print("Tokens:", tokens)

freq = FreqDist(tokens)
print("\nFrequency Distribution:")
print(freq.most_common(10))
```

## Question 8: Custom Annotator to Identify Proper Nouns

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

nltk.download('averaged_perceptron_tagger_eng') # Download the required tagger

text = "John lives in New York and works at Google"

tokens = word_tokenize(text)
tags = pos_tag(tokens)

print("Proper Nouns:")
for word, tag in tags:
    if tag == 'NNP':
        print(word)
```

## Question 9: Training Word2Vec Model using Gensim

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

dataset = [
 "Natural language processing enables computers to understand human language",
 "Word embeddings are a type of word representation that allows words with similar meaning",
 "Word2Vec is a popular word embedding technique",
 "Text preprocessing is a critical step before training word embeddings",
 "Tokenization and normalization help clean raw text for modeling"
]

processed = [word_tokenize(sentence.lower()) for sentence in dataset]

model = Word2Vec(sentences=processed, vector_size=50, window=3, min_count=1, workers=2)

print("Vector for 'language':")
print(model.wv['language'])
```

## Question 10: Steps to Analyze Customer Feedback using NLP

**Answer:**

2. **Data Cleaning** – Remove noise such as punctuation, stopwords, URLs, emojis, and special characters.

5. **Sentiment Analysis** – Classify reviews into positive, negative, or neutral sentiments.

6. **Topic Modeling** – Use algorithms such as LDA to discover common topics discussed by customers.

7. **Visualization & Insights** – Use charts, dashboards, or word clouds to present insights to stakeholders.
